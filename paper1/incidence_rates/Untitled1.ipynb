```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
results = pd.read_csv('incidence.csv')
```

```python
results
```

```python
for r in range(0, len(results.z1)):
    
    average_completeness = results.loc[r].N/results.loc[r].Ncomp
    redshift_path        = results.loc[r].z2 - results.loc[r].z1
    if redshift_path > 1.2 : redshift_path = 1.45
    total_redshift_path = 4*redshift_path
    corrected_redshift_path = total_redshift_path*average_completeness
    dndz_value       = results.loc[r].fp_com*results.loc[r].N/corrected_redshift_path
    dndz_value_error = results.loc[r].fp_com*results.loc[r].N/corrected_redshift_path/corrected_redshift_path
   
    dndz_value       = np.floor(dndz_value*10000.0)/10000
    dndz_value_error = np.floor(dndz_value_error*10000.0)/10000

    print(results.loc[r].z1,
          results.loc[r].z2,
          results.loc[r].ewlow,
          results.loc[r].ewhigh,
          results.loc[r].N,
          total_redshift_path,
          corrected_redshift_path,
          dndz_value, ' +/- ', dndz_value_error)
```
